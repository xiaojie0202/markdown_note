```python
'''
需求 ： 用户对物品类别的喜好细分降维
数据文件信息:
aisles.csv     商品所属具体物品类别
    aisle_id   类别ID
    aisle      类别名称

order_products__prior.csv   订单与商品信息
    order_id                订单ID
    product_id              商品ID
    add_to_cart_order       添加到购物车的订单
    reordered

orders.csv     用户的订单信息
    order_id   订单ID
    user_id    用户ID
    。。。
    
products.csv      商品信息
    product_id    商品ID
    product_name  商品名称
    aisle_id      商品所属类别ID
    department_id 商品所属店铺ID
'''
```

```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
prior = pd.read_csv('./data/order_products__prior.csv') # 订单与商品信息
orders = pd.read_csv('./data/orders.csv') # 用户订单信息
products = pd.read_csv('./data/products.csv') # 商品信息
aisles = pd.read_csv('./data/aisles.csv') # 商品类型信息
```

```python
_mg = pd.merge(prior, products, on=['product_id', 'product_id'])
_mg = pd.merge(_mg, orders, on=['order_id', 'order_id'])
meg = pd.merge(_mg, aisles, on=['aisle_id', 'aisle_id'])
```

```python
meg.head(10)
```

```python
cross = pd.crosstab(meg['user_id'], meg['aisle'])
```

```python
cross.head(10)
```

```python
pca = PCA(n_components=0.9)
```

```python
data = pca.fit_transform(cross)
```
